```python
import threading
import queue
import time
```

```python
mailbox = queue.Queue(maxsize=2)   # mailbox with capacity 2 messages
```

```python
def mailbox_producer():
    for i in range(5):
        msg = f"MSG-{i}"
        print("Producer sending:", msg, flush=True)
        mailbox.put(msg)           # blocks if mailbox is full
        time.sleep(0.3)
```

```python
def mailbox_consumer():
    for _ in range(5):
        msg = mailbox.get()        # blocks if mailbox is empty
        print("Consumer received:", msg, flush=True)
        time.sleep(0.3)
```

```python
t1 = threading.Thread(target=mailbox_producer)
t2 = threading.Thread(target=mailbox_consumer)

t1.start()
t2.start()
t1.join()
t2.join()
```
