```python
from multiprocessing.dummy import Process
from multiprocessing import Array, Value, Lock
import time
import random
```

```python
BUFFER_SIZE = 5
TOTAL_ITEMS = 10

buffer = Array('i', BUFFER_SIZE) 
count = Value('i', 0)            
in_index = Value('i', 0)         
out_index = Value('i', 0)        

#lock = Lock()
```

```python
def producer():
    for item in range(1, TOTAL_ITEMS + 1):
        produced = random.randint(10, 100)

        while True:
            if count.value < BUFFER_SIZE:
                    buffer[in_index.value] = produced
                    print(f"Producer -> Inserted {produced} at position {in_index.value}")

                    in_index.value = (in_index.value + 1) % BUFFER_SIZE
                    count.value += 1
                    break
            # with lock:
            #     if count.value < BUFFER_SIZE:
            #         buffer[in_index.value] = produced
            #         print(f"Producer -> Inserted {produced} at position {in_index.value}")

            #         in_index.value = (in_index.value + 1) % BUFFER_SIZE
            #         count.value += 1
            #         break

            time.sleep(0.1)

        time.sleep(random.uniform(0.2, 0.5))


def consumer():
    for _ in range(TOTAL_ITEMS):
        while True:
            if count.value > 0:
                    value = buffer[out_index.value]
                    print(f"Consumer <- Removed  {value} from position {out_index.value}")

                    out_index.value = (out_index.value + 1) % BUFFER_SIZE
                    count.value -= 1
                    break
            # with lock:
            #     if count.value > 0:
            #         value = buffer[out_index.value]
            #         print(f"Consumer <- Removed  {value} from position {out_index.value}")

            #         out_index.value = (out_index.value + 1) % BUFFER_SIZE
            #         count.value -= 1
            #         break

            time.sleep(0.1)

        time.sleep(random.uniform(0.2, 0.5))
```

```python
if __name__ == "__main__":
    p = Process(target=producer)
    c = Process(target=consumer)

    p.start()
    c.start()

    p.join()
    c.join()

    print("Done! Producer created 10 items and Consumer consumed 10 items.")
```
