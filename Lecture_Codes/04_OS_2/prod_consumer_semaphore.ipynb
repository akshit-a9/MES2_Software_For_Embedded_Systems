```python
import threading
import queue
import time
import random
```

```python
BUFFER_SIZE = 5
buffer = queue.Queue(maxsize=BUFFER_SIZE)
```

```python
empty_slots = threading.Semaphore(BUFFER_SIZE)
full_slots = threading.Semaphore(0)
mutex = threading.Semaphore(1)
```

```python
def camera():
    frame_id = 0
    while frame_id < 10:
        time.sleep(random.uniform(0.5, 1.2))

        empty_slots.acquire()
        mutex.acquire()

        buffer.put(f"Frame-{frame_id}")
        print(f"[Camera] Captured Frame-{frame_id}")

        mutex.release()
        full_slots.release()

        frame_id += 1
```

```python
def satellite():
    for _ in range(10):
        full_slots.acquire()
        mutex.acquire()

        frame = buffer.get()
        print(f"    [Satellite] Processing {frame}")

        mutex.release()
        empty_slots.release()

        time.sleep(random.uniform(0.8, 1.5))
```

```python
producer = threading.Thread(target=camera)
consumer = threading.Thread(target=satellite)

producer.start()
consumer.start()

producer.join()
consumer.join()

print("\nSimulation complete.")
```
